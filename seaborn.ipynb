<h1 style="text-align:center; color:red">Seaborn</h1>

<h2 style="color:blue">Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn is particularly useful for visualizing complex datasets and for creating aesthetically pleasing and informative plots with minimal code.</h2>

```python
import seaborn as sns ## importing seaborn for plotting
import matplotlib.pyplot as plt ## importing matplotlib for plotting
import pandas as pd ## importing pandas for dataframe
import numpy  ## importing numpy for array
```

```python
df =sns.load_dataset('tips') ## loading dataset 'tips'. It is an inbuilt dataset.
df.head() ## prints top 5 rows
```

```python
df.columns=['total_bill','tip','gender','smoker','day','time','size'] ## renaming sex to gender
```

<h2 style="text-align:center;">ScatterPlot</h2>

<h3>A scatter plot is a type of data visualization that displays values for two variables for a set of data using Cartesian coordinates. Each point on the scatter plot represents an observation in the dataset, with the position along the x-axis and y-axis representing the values for the two variables.</h3>

```python
sns.scatterplot(x='total_bill',y='tip',data=df) ## plotting gaph with scatterplot
plt.show()
```

<h2 style="text-align:center">Relational Plot (relplot)</h3>

<h3>A relational plot is a type of data visualization used to show the relationships between variables. In Seaborn, relplot is a versatile function for creating relational plots. It combines different kinds of plots, such as scatter plots and line plots, allowing for complex visualizations.</h3>

<h3>Key Features:-

1) Flexibility: Can create scatter plots and line plots.

2) Faceting: Supports faceting by additional categorical variables using the col and row parameters.

3) Hue, Size, Style: Allows mapping of additional dimensions using color (hue), size (size), and marker style (style).</h3>

```python
sns.relplot(x="total_bill",y="tip",data=df) #relationtional plot by creates scatter plot by-default
# plotting graph with relational plot for "total_bill" column.
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="line") # we can plot graph by another plot by kind parameter.
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker") #we can pass x-axis, y-axis, hue(differentiate by colour), style(differ by structures)
```

```python
df
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="smoker",col="time") #we can differ them by columnwise graph by passing value to col parameter.
plt.show()
#total columns = number of unique items.
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",col="day") #passing day to col parameter.
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="gender",style="time",col="gender")
plt.show()
```

```python
female=df[df["gender"]=="Female"] ## filtering data
```

```python
sns.relplot(x="tip",y="total_bill",data=female,hue="smoker",row="day") ## we can differ columns by rows by passing column name to row parameter.
#plotting graph for females (smoker and non smoker), daywise by passing x-axis for tip and y-axis for total bill.
```

<h3>A count plot is a type of bar plot that displays the count of observations in each category of a categorical variable. It helps visualize the frequency distribution of categories in a dataset.</h3>

```python
sns.countplot(x="day",data=df,hue="smoker")  ## plotting graph using countplot according to day
plt.show()
```

<h2 style="text-align:center">Categorical Plot</h2>

<h3>Catplot (short for "categorical plot") is a high-level interface in Seaborn that allows users to create a variety of categorical plots, such as bar plots, box plots, violin plots, and more, using a single function. It is designed to simplify the process of visualizing categorical data.</h3>

<h3>Key Features:-

1) Versatile: Supports multiple plot types (bar, box, violin, point, etc.) via the kind parameter.

2) Faceting: Easily creates multiple subplots using col and row for comparison.

3) Automatic Aggregation: Automatically summarizes data for the selected plot type.

4) Customizable: Offers various options for aesthetic customization.

5) Pandas Integration: Works seamlessly with pandas DataFrames for easy data manipulation.</h3>

<h3 style="text-align:center">||<br>

<h3>Swarm:- A swarm plot is a type of data visualization that displays individual data points for a categorical variable, showing their distribution while avoiding overlap.</h3>

```python
sns.catplot(x="day",y="total_bill",data=df,kind="swarm") ## we can change type of plot in catplot by passing value to kind parameter.
# plotting graph by passing day to x-axis and total_bill to y-axis
```

<h3>Violin:- A violin plot displays the distribution of a continuous variable for different categories, showing both the density of the data and summary statistics.</h3>

```python
sns.catplot(x="day",y="total_bill",data=df,kind="violin") ## violin type graph
```

<h3>In Seaborn, you can create a bar plot using the catplot function by setting the kind parameter to 'bar'. This allows you to visualize the average value of a continuous variable for different categories.</h3>

```python
sns.catplot(x="day",y="total_bill",data=df,kind="bar") ## plotting graph by barplot in catplot and similar to barplot in matplolib
```

<h3>A boxen plot is a variation of a box plot designed to provide a more detailed view of the distribution of data, especially for larger datasets. It's particularly useful for visualizing data with a lot of outliers or for comparing distributions across multiple categories.</h3>

```python
sns.catplot(x="day",y="total_bill",data=df,kind="boxen") ## boxen is similar to boxplot in matplotlib
```

<h3>A point plot is a type of categorical plot that visualizes the relationship between a categorical variable and a continuous variable by displaying summary statistics (typically the mean) as points, often with error bars to indicate variability.</h3>

```python
sns.catplot(x="day",y="total_bill",data=df,kind="point") ##pointplot 
```
